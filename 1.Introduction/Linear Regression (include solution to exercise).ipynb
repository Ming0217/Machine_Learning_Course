# 机器学习入门之线性回归
你所在的公司在电视上做产品广告, 收集到了电视广告投入x(以百万为单位)与产品销售量y(以亿为单位)的数据. 你作为公司的数据科学家, 希望通过分析这些数据, 了解电视广告投入x(以百万为单位)与产品销售量y的关系. 

假设x与y的之间的关系是线性的, 也就是说 y = ax + b. 通过线性回归(Linear Regression), 我们就可以得知 a 和 b 的值. 于是我们在未来做规划的时候, 通过电视广告投入x, 就可以预测产品销售量y, 从而可以提前做好生产和物流, 仓储的规划. 为客户提供更好的服务.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("data/Advertising.csv")
```

```python
data.head(5)
```

```python
data.columns
```

## 通过数据可视化分析数据

```python
plt.figure(figsize=(16, 8))
plt.scatter(data['TV'], data['sales'], c ='blue')
plt.xlabel("Money spent on TV ads")
plt.ylabel("Sales")
plt.show()
```

## 训练线性回归模型

*Note: when -1 is passed into the reshape method, it means that the size of the dimension, for which you passed -1, is being inferred based on the total size of the array and the other known dimension*

```python
X = data['TV'].values.reshape(-1,1)
y = data['sales'].values.reshape(-1,1)

reg = LinearRegression()
reg.fit(X, y)
```

```python
print('a = {:.5}'.format(reg.coef_[0][0]))
print('b = {:.5}'.format(reg.intercept_[0]))

print("线性模型为: Y = {:.5}X + {:.5} ".format(reg.coef_[0][0], reg.intercept_[0]))
```

## 可视化训练好的线性回归模型

```python
predictions = reg.predict(X)

plt.figure(figsize=(16, 8))
plt.scatter(data['TV'], data['sales'], c ='blue')
plt.plot(data['TV'], predictions,c ='red', linewidth=2)
plt.xlabel("Money spent on TV ads")
plt.ylabel("Sales")
plt.show()
```

```python
# 做预测
假设公司希望在下一个季度投一亿元的电视广告, 那么预期的产品销量会是多少呢?
```

```python
predictions = reg.predict([[100])
print('投入一亿元的电视广告, 预计的销售量为{:.5}亿'.format( predictions[0][0]) )
```

```python
data2 = pd.read_csv("exercise/height.vs.temperature.csv")
```

```python
data2
```

### Visualization of raw data

```python
plt.figure(figsize=(16, 8))
plt.scatter(data2['height'], data2['temperature'], c ='blue')
plt.xlabel("height")
plt.ylabel("temperature")
plt.show()
```

### Training

```python
X2 = data2['height'].values.reshape(-1,1)
y2 = data2['temperature'].values.reshape(-1,1)

model = LinearRegression()
model.fit(X2, y2)
```

### Model 

```python
print('a = {:.5}'.format(model.coef_[0][0]))
print('b = {:.5}'.format(model.intercept_[0]))

print("线性模型为: Y = {:.5}X + {:.5} ".format(model.coef_[0][0], model.intercept_[0]))
```

### Visualization of Model

```python
predictions = model.predict(X2)

plt.figure(figsize=(16, 8))
plt.scatter(data2['height'], data2['temperature'], c ='blue')
plt.plot(data2['height'], predictions,c ='red', linewidth=2)
plt.xlabel("Height")
plt.ylabel("Temperature")
plt.show()
```

### Prediction

```python
predictions = model.predict([[8000]])
print('海拔8000米的温度预计为{:.5}度'.format( predictions[0][0]) )
```
